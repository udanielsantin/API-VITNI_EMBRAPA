```python
import requests
```

```python
BASE_URL = "https://api-vitni-embrapa.onrender.com"
```

```python
USERNAME = "vapor"
PASSWORD = "max"
```

```python
res = requests.post(f"{BASE_URL}/register", json={
    "username": USERNAME,
    "password": PASSWORD
})
print(res.status_code)
print(res.json())
```

```python
res = requests.post(f"{BASE_URL}/token", data={
    "grant_type": "password",
    "username": USERNAME,
    "password": PASSWORD
})
print(res.status_code)
print(res.json())
```

```python
token = res.json()["access_token"]

headers = {"Authorization": f"Bearer {token}"}
```

```python
# Exemplo de requisição autenticada
# PRODUCAO
data = requests.get(f"{BASE_URL}/scrape/producao?ano=2021", headers=headers)
print(data.json())
```

```python
# Processamento
print(requests.get(f"{BASE_URL}/scrape/processamento/americanas_hibridas?ano=2020", headers=headers).json())
print(requests.get(f"{BASE_URL}/scrape/processamento/sem_classificacao?ano=2020", headers=headers).json())
print(requests.get(f"{BASE_URL}/scrape/processamento/uvas_de_mesa?ano=2020", headers=headers).json())
print(requests.get(f"{BASE_URL}/scrape/processamento/viniferas?ano=2020", headers=headers).json())
```
